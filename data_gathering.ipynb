```python
all_jobs = []

for i in range(1,30):

    response = requests.get(f"https://www.104.com.tw/jobs/search/api/jobs?jobsource=joblist_search&keyword=%E6%A9%9F%E5%99%A8%E5%AD%B8%E7%BF%92&mode=s&order=15&page={i}&pagesize=20", headers= my_headers)
    data = response.json()
    info = data['data']
    #'jobNameSnippet'
    all_jobs = [d for d in info]
    all_jobs.extend(info)
    time.sleep(2)
print(all_jobs)
```

```python
cleaned_data= []
for job in all_jobs:
    raw_title = job.get("jobName", "")
    clean_title = raw_title.replace('[[[', '').replace(']]]', '')
    low = job.get('salaryLow', 0)
    high = job.get('salaryhigh', 0)

    cleaned_data.append({
        'company': job.get("custName"),
        "title": clean_title,
        'description': job.get('description'),
        'low_salary': low,
        "high_salary" : high,
        
        "link": job.get('link', {}).get('job'),
        "address": job.get("jobAddrNoDesc"),
        'pcSkills': job.get("pcSkills")
    })
    
    #print(cleaned_data)
df = pd.DataFrame(cleaned_data)
df.drop_duplicates(subset = ['jobName', "custName"], keep = 'first',inplace = True)
df.head(5)
```

```python
def avg_salary(row):
    low = row['low_salary']
    high = row['high_salary']
    if low == 0 and high  == 0: #negaotiable 
        return None
    elif (low != 0 and high == 0 )or (low== 0 and high != 0):
        return max(low,high)
    else:
        return (low+ high)/2
df['avg_salary'] = df.apply(avg_salary, axis = 1)
print(df[['title', 'low_salary', 'high_salary', 'avg_salary']].head(10))
```

```python
def extract_skills(skill_lst):
    if not skill_lst:
        return "NA"
    
    skills = [x['description'] for x in skill_lst]
    return ", ".join(skills)

df['skills_tags'] = df['pcSkills'].apply(extract_skills)
print(df[['title', 'pcSkills']].head())
```
